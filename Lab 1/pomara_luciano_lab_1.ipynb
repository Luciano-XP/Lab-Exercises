1. See the list below. Please write a program that will tell you what the longest word is in the list. Package your program inside of a function.  

```python
#The list mentioned in the question
list = ["apple", "chestnut", "gargoyle", "pandas", "sheep", "raptor"]

def longest_word(words):
    def sort_by_size(x):    #https://www.w3schools.com/python/ref_list_sort.asp
        return len(x)
    words.sort(reverse=True, key=sort_by_size) #https://www.w3schools.com/python/ref_list_sort.asp
    return words[0]
        
print(longest_word(list))
```

```python
import random 
word_txt = "words.txt"

def word_generator():
        filename = word_txt
        f = open(filename, "r", encoding="utf8")
        words = f.readlines()
        f.close()
        return random.choice(words).strip()

def seven_random_words():
        random_words = []
        while len(random_words)<7:
                random_words.append(word_generator())
        return random_words

x = seven_random_words()
print(x)
print(f"The longest word is: {longest_word(x)}")
```

3. Pretend that you're going to be leading a review session on dictionaries. Build a program that uses dictionaries and incorporates at least 4 built-in methods. [See full list on W3Schools](https://www.w3schools.com/python/python_ref_dictionary.asp) 

```python
the_dictionary = {"apple": 2,
                  "banana": 3,
                  "orange": 4,
                  "pear": 5}

def dictionary_review(dict):
    print("This function will begin by showing off the pop method.\nThe pop() method removes the specified key and its value from a dictionary and\nthen returns the value.")
    print("To show this, we will use pop() with the \"apple\" key.")
    print(f"Before: {dict}")
    popped = dict.pop("apple")
    print(f"Popped value using dict.pop(\"apple\"): {popped}")
    print(f"After: {dict}")

    print("\nThe next method to go over will be the update() method.\nThis method adds a specified key-value pair to the dictionary.")
    print("Mango will be added to the dictionary with a key of 6 by using dict.update({\"mango\": 6}).")
    dict.update({"mango": 6})
    print(f"After: {dict}")

    print("\nNext will be the popitem() method.\nThis method differs from the previous pop() method as it removes\nthe last item added to the dictionary.\nThe method returns the removed key-value pair as a tuple.")
    print("No parameter needs to be given for popitem(). By using it here,\nit will effectively undo the addition of the key \"mango\".")
    print("This will be done with the line dict.popitem()")
    popped_item = dict.popitem()
    print(f"After: {dict}")
    print(f"The popped key-value pair, returned as a tuple: {popped_item}\n")

    print("Finally, we have the keys() function. This function is\nvery simple and convenient. It returns a view object showing a list of all the keys\nin the dictionary. This can be especially useful for iterating through\na dictionary.")
    print("To use it, we will use the line dict.keys()")
    keys_list = dict.keys()
    print(f"Returned list of keys: {keys_list}")

dictionary_review(the_dictionary)
```
